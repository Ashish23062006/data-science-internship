```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv("Dataset .csv")
```

```python
#clean data
df = df.dropna(subset=['Restaurant Name', 'Cuisines', 'Votes', 'Aggregate rating', 'Price range'])
```

```python
# top rated restaurants
top_rated = df[df['Aggregate rating'] >=4.5]
top_rated[['Restaurant Name', 'Aggregate rating', 'Cuisines']].head(10)
```

```python
#most common cuisines
df['Cuisines'].value_counts().head(10).plot(kind='barh', figsize=(9,6), color= 'teal')

plt.title("Top 10 Most Common Cuisines")
plt.xlabel("Number of Restaurant")
plt.show()
```

```python
#votes vs ratings
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Aggregate rating', y='Votes', hue='Price range', alpha=0.6)
plt.title("Votes vs Ratings with Price Range")
plt.show()
```

```python
#price Range distribution
df['Price range'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(7,7))

plt.title("Distritution of Price Range")
plt.ylabel("")
plt.show()
```
