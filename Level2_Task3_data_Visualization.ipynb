```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load data
df = pd.read_csv("Dataset .csv")
```

```python
#clean data
df = df.dropna(subset=['Aggregate rating', 'Votes', 'Price range', 'Cuisines',])
```

```python
#rating distribution histogram
plt.figure(figsize=(9,6))
sns.histplot(df['Aggregate rating'], bins=10, kde=True, color='green')

plt.title("Distribution of Aggregate Ratings")
plt.xlabel("Rating")
plt.ylabel("count")
plt.show()
```

```python
# boxplot- price range vs rating 
n_colors = df['Price range'].nunique()
pastel_colors = sns.color_palette("pastel", n_colors)
plt.figure(figsize=(7,5))
sns.boxplot(x='Price range', y= 'Aggregate rating', data=df, hue='Price range', palette=pastel_colors, legend=False)
plt.title("Price Range vs Rating")
plt.xlabel("Price Range")
plt.ylabel("Aggregate Rating")
plt.tight_layout()
plt.show()
```

```python
#votes vs rating trend
plt.figure(figsize=(8,5))
sns.regplot(x='Aggregate rating', y='Votes', data=df, scatter_kws={'alpha':0.5})
plt.title("Votes vs Rating Trend")
plt.show()
```

```python
#top 5 cuisines by average rating 
top_cuisines = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(5)
top_cuisines.plot(kind='bar', figsize=(6,4), title='Top 5 Cuisines by Avg Rating', color='orange')
plt.ylabel("Average Rating")
plt.xticks(rotation=45, ha= 'right')
plt.show()
```

```python
df['Rating Class'] = df['Aggregate rating'].apply(lambda x: 'High' if x >=4 else 'Low')
print(df[['Cuisines', 'Aggregate rating', 'Rating Class']].head())
```
