```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("Dataset .csv")
```

```python
#basic state for numerical columns
df.describe()
```

```python
print("Mean: \n",df.mean(numeric_only=True))
print("Median: \n",df.median(numeric_only=True))
print("Standard Deviation: \n",df.std(numeric_only=True))
```

```python
#country code distribution
country_counts = df['Country Code'].value_counts()
print(country_counts)

#optional plot
country_counts.plot(kind='bar',figsize=(8,4), title='Country Code Distribution')
plt.xlabel("Country Code")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
#city distribution
city_counts = df['City'].value_counts().head(10)
print(city_counts)

city_counts.plot(kind='barh',figsize=(10,5), title='Top 10 Cities with Most Restaurants')
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.show()
```

```python
#cuisines distribution 
cuisine_counts = df['Cuisines'].value_counts().head(10)
print(cuisine_counts)

cuisine_counts.plot(kind='bar',figsize=(10,6), title='Top 10 Cuisines')
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')
plt.show()
```
